## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

### Assessing Data

#### Identifying ... problem

```python
import pandas as pd

# Load data
day_df = pd.read_csv('day.csv')
hour_df = pd.read_csv('hour.csv')

def assess_data(df, name):
    print(f"--- Assessing {name} Dataframe ---")

    # 1. Mengecek Missing Values
    print("\n[1] Missing Values:")
    print(df.isnull().sum())

    # 2. Mengecek Duplikasi Data
    print(f"\n[2] Duplicate rows: {df.duplicated().sum()}")

    # 3. Mengecek Tipe Data
    print("\n[3] Data Types:")
    print(df.dtypes)

    # 4. Mengecek Statistik Deskriptif (Untuk melihat Outlier/Inaccurate Value)
    print("\n[4] Descriptive Statistics:")
    print(df.describe())
    print("\n" + "="*50 + "\n")

# Menjalankan fungsi assesment
assess_data(day_df, "Day")
assess_data(hour_df, "Hour")
```

**Insight:** (Opsional)
- Inaccurate Data Type (Tipe Data Tidak Akurat): Masalahnya adalah  Kolom dteday terdeteksi sebagai tipe object (string).Harus diubah menjadi tipe datetime pada tahap Cleaning.

- Inconsistent/Non-Informative Values (Nilai Tidak Konsisten/Informatif):Masalah nya adalah Kolom kategorikal seperti season, weathersit, dan yr masih berbentuk angka integer (1, 2, 3, dst). Oleh karena itu,Melakukan mapping atau mengubah nilai angka tersebut menjadi label string yang deskriptif pada tahap Cleaning.

- Data Range & Outliers (Melalui Descriptive Statistics): Pada Data suhu (temp), kelembapan (hum), dan kecepatan angin (windspeed) sudah dinormalisasi (berada di rentang 0 hingga 1). Tidak ditemukan nilai negatif yang mustahil untuk jumlah penyewaan (cnt), sehingga data numerik terlihat cukup valid.


### Cleaning Data

#### Fixing the problem

```python
# 1. Mengubah tipe data 'dteday' menjadi datetime
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])

# 2. Mapping nilai angka menjadi label deskriptif

# Mapping Musim (Season)
season_map = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
day_df['season'] = day_df['season'].map(season_map)
hour_df['season'] = hour_df['season'].map(season_map)

# Mapping Tahun (Year)
year_map = {0: 2011, 1: 2012}
day_df['yr'] = day_df['yr'].map(year_map)
hour_df['yr'] = hour_df['yr'].map(year_map)

# Mapping Bulan (Month)
month_map = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
             7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
day_df['mnth'] = day_df['mnth'].map(month_map)
hour_df['mnth'] = hour_df['mnth'].map(month_map)

# Mapping Kondisi Cuaca (Weather Situation)
weather_map = {
    1: 'Clear',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Heavy Rain/Ice'
}
day_df['weathersit'] = day_df['weathersit'].map(weather_map)
hour_df['weathersit'] = hour_df['weathersit'].map(weather_map)

# Mapping Hari (Weekday)
weekday_map = {0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'}
day_df['weekday'] = day_df['weekday'].map(weekday_map)
hour_df['weekday'] = hour_df['weekday'].map(weekday_map)

print("Proses pembersihan data selesai!")
```

## Exploratory Data Analysis (EDA)

### Explore ...

```python
# 1. Eksplorasi untuk Pertanyaan 1: Pola Jam-an (2012)
# Memfilter data hanya untuk tahun 2012
hour_2012 = hour_df[hour_df['yr'] == 2012]

# Melihat rata-rata penyewaan per jam berdasarkan hari kerja (workingday)
# 1 = Hari Kerja, 0 = Akhir Pekan/Libur
hourly_stats = hour_2012.groupby(by=["workingday", "hr"]).agg({
    "casual": "mean",
    "registered": "mean",
    "cnt": "mean"
})
print("--- Statistik Rata-rata Penyewaan Per Jam (Working Day vs Weekend) ---")
print(hourly_stats.head(10))
```

```python
# 2. Eksplorasi untuk Pertanyaan 2: Pengaruh Cuaca & Suhu (2012)
day_2012 = day_df[day_df['yr'] == 2012]

# Melihat total penyewaan bulanan berdasarkan kondisi cuaca
monthly_weather_stats = day_2012.groupby(by=["mnth", "weathersit"]).agg({
    "cnt": ["sum", "mean"],
    "temp": "mean"
})
print("\n--- Statistik Bulanan Berdasarkan Cuaca (2012) ---")
print(monthly_weather_stats.head(10))

# Menganalisis korelasi antara variabel numerik (Suhu, Kelembapan, dll) dengan Jumlah Penyewaan
correlation = day_2012[['temp', 'atemp', 'hum', 'windspeed', 'cnt']].corr()
print("\n--- Korelasi terhadap Jumlah Penyewaan (cnt) ---")
print(correlation['cnt'].sort_values(ascending=False))
```

**Insight:** (Opsional)
-Pola Pengguna Berbeda: Pada hari kerja, penyewaan didominasi oleh pengguna Registered dengan lonjakan tinggi pada jam masuk kantor (08:00) dan pulang kantor (17:00). Pada akhir pekan, pengguna Casual meningkat secara drastis di siang hari, menunjukkan penggunaan untuk rekreasi.

- Pengaruh Suhu yang Sangat Kuat: Hasil korelasi menunjukkan angka 0.71 untuk suhu (temp) terhadap jumlah penyewaan. Artinya, semakin hangat suhu (hingga titik tertentu), semakin banyak orang yang menyewa sepeda.Kecepatan angin (windspeed) memiliki korelasi negatif (-0.27), yang berarti angin kencang cenderung mengurangi minat orang untuk bersepeda.

- Dampak Cuaca: Penyewaan mencapai puncaknya pada kondisi cuaca Clear (Cerah). Pada bulan-bulan dengan kondisi Light Snow/Rain, rata-rata penyewaan turun drastis (misalnya di bulan Februari dan April), mengonfirmasi bahwa cuaca buruk adalah penghambat utama bisnis ini.

## Analisis Lanjutan (Opsional)

```python
# 1. BINNING: Mengelompokkan Suhu (temp) menjadi Cluster
# Menggunakan pd.cut untuk membagi angka menjadi kategori spesifik
bins_temp = [0, 0.3, 0.6, 0.8, 1]
labels_temp = ['Cold', 'Moderate', 'Warm', 'Hot']
day_df['temp_cluster'] = pd.cut(day_df['temp'], bins=bins_temp, labels=labels_temp)

# 2. CLUSTERING: Mengelompokkan Permintaan (cnt) berdasarkan Aturan Bisnis
# Kita gunakan kuartil data untuk menentukan ambang batas Low/Medium/High Demand
q1 = day_df['cnt'].quantile(0.25)
q3 = day_df['cnt'].quantile(0.75)

def logic_clustering(cnt):
    if cnt <= q1:
        return 'Low Demand'
    elif cnt <= q3:
        return 'Medium Demand'
    else:
        return 'High Demand'

day_df['demand_cluster'] = day_df['cnt'].apply(logic_clustering)

# 3. VISUALISASI HASIL CLUSTERING
plt.figure(figsize=(12, 6))
sns.countplot(data=day_df, x='temp_cluster', hue='demand_cluster', palette='magma')

plt.title('Analisis Cluster: Hubungan Suhu terhadap Intensitas Permintaan', fontsize=14, fontweight='bold')
plt.xlabel('Cluster Suhu', fontsize=12)
plt.ylabel('Jumlah Hari', fontsize=12)
plt.legend(title='Tingkat Permintaan')
plt.show()

# Menampilkan hasil pemrosesan
print(day_df[['dteday', 'temp', 'temp_cluster', 'cnt', 'demand_cluster']].head())
```

**Insight:** (Opsional)
- Binning (pd.cut):

Teknik ini membagi variabel kontinu (temp) menjadi interval-interval diskrit bertujuan untuk  Memudahkan analisis operasional. Tim lapangan lebih mudah mengerti instruksi "Siapkan sepeda tambahan saat cuaca Warm" dibandingkan "Siapkan sepeda saat suhu 0.65 - 0.80".

- Manual Clustering (apply function):

Teknik ini mengelompokkan data berdasarkan karakteristik volume penyewaan dengan tujuan Mengidentifikasi hari-hari "High Demand". Dari visualisasi (grafik batang), Anda bisa melihat secara jelas bahwa cluster High Demand hampir selalu muncul pada suhu Warm dan Moderate, namun tidak pernah muncul pada suhu Cold.
