## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

```python
print("Peningkatan penyewaan Musim Panas (Fall - season=3) vs Musim Dingin (Spring - season=1)\n")

# Rata-rata penyewaan per musim
season_avg = df_day.groupby('season', observed=True)['cnt'].mean().round(2)
print("Rata-rata penyewaan per musim:")
print(season_avg)

fall = season_avg[3]
spring = season_avg[1]
peningkatan_persen = ((fall - spring) / spring) * 100

print(f"\nMusim Dingin (Spring) : {spring:,.0f} sepeda/hari")
print(f"Musim Panas (Fall)    : {fall:,.0f} sepeda/hari")
print(f"Peningkatan           : {peningkatan_persen:.1f}%")

# Faktor cuaca (hanya kolom numerik)
weather_by_season = df_day.groupby('season', observed=True)[['temp', 'atemp', 'hum', 'windspeed']].mean().round(3)
print("\nRata-rata Faktor Cuaca per Musim:")
display(weather_by_season)

```

```python
print("\n" + "="*60)
print("PERTANYAAN 2: Jam Puncak Registered Users\n")

hourly_reg = df_hour.groupby(['workingday', 'hr'], observed=True)['registered'].mean().unstack(0)
hourly_reg.columns = ['Weekend/Holiday', 'Working Day']

print("Rata-rata Registered Users per Jam:")
display(hourly_reg.round(1))

# Jam puncak
peak_working = hourly_reg['Working Day'].idxmax()
peak_libur = hourly_reg['Weekend/Holiday'].idxmax()

print(f"\nJam puncak Hari Kerja   → Jam {peak_working:02d} ({hourly_reg['Working Day'].max():.0f} orang)")
print(f"Jam puncak Hari Libur   → Jam {peak_libur:02d} ({hourly_reg['Weekend/Holiday'].max():.0f} orang)")

# Potensi Revenue 2012
df_2012 = df_hour[df_hour['yr'] == 1]

peak_hour_data = df_2012[(df_2012['hr'] == peak_working) & (df_2012['workingday'] == 1)]
avg_peak = peak_hour_data['registered'].mean()

additional = avg_peak * 0.20
revenue_est = additional * 365 * 3   # asumsi $3 per jam

print(f"\nPotensi tambahan armada (20%): {additional:.0f} sepeda")
print(f"Estimasi tambahan revenue (2012): ${revenue_est:,.0f}")
```

```python
print(" RFM Analysis lanjutan (Segmentasi Hari)\n")

# Pastikan dteday bertipe datetime
df_day['dteday'] = pd.to_datetime(df_day['dteday'])

# Buat RFM table
rfm = df_day.copy()

# Recency
max_date = rfm['dteday'].max()
rfm['Recency'] = (max_date - rfm['dteday']).dt.days

# Frequency & Monetary (menggunakan registered users)
rfm['Frequency'] = rfm['registered']
rfm['Monetary'] = rfm['registered'] * 3        # asumsi harga sewa $3 per transaksi

# RFM Scoring
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'], 5, labels=[1,2,3,4,5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])

rfm['RFM_Score'] = rfm['R_Score'].astype(int) + rfm['F_Score'].astype(int) + rfm['M_Score'].astype(int)

# Segmentasi
def rfm_segment(score):
    if score >= 12: return 'Champion'
    elif score >= 9: return 'Loyal Customer'
    elif score >= 6: return 'Potential'
    else: return 'At Risk'

rfm['Segment'] = rfm['RFM_Score'].apply(rfm_segment)

print("RFM Summary:")
display(rfm[['dteday', 'Recency', 'Frequency', 'Monetary', 'RFM_Score', 'Segment']].head())

print("\nDistribusi Segment:")
print(rfm['Segment'].value_counts())

```

```python
import pandas as pd

# Load data
df_hour = pd.read_csv('hour.csv')

# 1. Rata-rata registered per jam pada hari kerja vs hari libur
reg_by_hour = df_hour.groupby(['workingday', 'hr'])['registered'].mean().round(2).reset_index()

# Pivot supaya lebih mudah dibaca
pivot_reg = reg_by_hour.pivot(index='hr', columns='workingday', values='registered')
pivot_reg.columns = ['Hari Libur (0)', 'Hari Kerja (1)']
pivot_reg = pivot_reg.round(2)

# Jam puncak pada hari kerja
peak_working = pivot_reg.sort_values('Hari Kerja (1)', ascending=False).head(10)

print("=== JAM PUNCAK REGISTERED USERS (Hari Kerja) ===")
print(peak_working[['Hari Kerja (1)']])

# Ambil 5 jam puncak tertinggi di hari kerja
top5 = pivot_reg.nlargest(5, 'Hari Kerja (1)')

result_q2 = pd.DataFrame({
    'Jam (hr)': top5.index,
    'Rata-rata Registered Hari Kerja': top5['Hari Kerja (1)'].values,
    'Rata-rata Registered Hari Libur': top5['Hari Libur (0)'].values,
    'Selisih (Kerja - Libur)': (top5['Hari Kerja (1)'] - top5['Hari Libur (0)']).values
})

print("\n=== DATAFRAME HASIL PERTANYAAN 2 ===")
print(result_q2.to_string(index=False))

# Hitung potensi peningkatan armada 20% tahun 2012
df_2012 = df_hour[df_hour['yr'] == 1]

peak_hours = [7, 8, 9, 16, 17, 18]   # jam puncak pagi dan sore

total_registered_peak_2012 = df_2012[
    (df_2012['hr'].isin(peak_hours)) &
    (df_2012['workingday'] == 1)
]['registered'].sum()

tambahan_20 = int(total_registered_peak_2012 * 0.20)

print(f"\nTotal Registered di jam puncak tahun 2012          : {total_registered_peak_2012:,}")
print(f"Potensi tambahan penyewaan (+20% armada)           : {tambahan_20:,}")

# Simpan ke Excel
with pd.ExcelWriter('Hasil_Pertanyaan_2.xlsx') as writer:
    result_q2.to_excel(writer, sheet_name='Jam_Puncak', index=False)
    pivot_reg.to_excel(writer, sheet_name='All_Hours')

print("\n✅ DataFrame berhasil dibuat dan disimpan ke 'Hasil_Pertanyaan_2.xlsx'")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# --- KONFIGURASI HALAMAN ---
st.set_page_config(page_title="Bike Sharing Dashboard", layout="wide")

# --- LOAD DATA ---
@st.cache_data
def load_data():
    # Mengasumsikan all_data.csv berisi data gabungan (hour & day)
    df = pd.read_csv("all_data.csv")
    df['dteday'] = pd.to_datetime(df['dteday'])
    return df

df_all = load_data()

# --- HEADER ---
st.title("🚲 Bike Sharing Analytics Dashboard")
st.markdown("Dashboard ini menampilkan hasil analisis mendalam tentang tren penyewaan, perilaku pengguna, dan segmentasi pelanggan.")

# --- SIDEBAR FILTER ---
st.sidebar.header("Filter Global")
year_option = st.sidebar.selectbox("Pilih Tahun:", options=[2011, 2012])
yr_filter = 0 if year_option == 2011 else 1
df_filtered = df_all[df_all['yr'] == yr_filter]

# --- LAYOUT TABS ---
tab1, tab2, tab3 = st.tabs(["📊 Analisis Musim & Cuaca", "⏰ Registered Users & Korelasi", "🎯 Segmentasi RFM"])

# --- TAB 1: ANALISIS MUSIM & CUACA ---
with tab1:
    st.header(f"Analisis Musim & Cuaca Tahun {year_option}")

    col1, col2 = st.columns(2)

    with col1:
        st.subheader("Rata-rata Penyewaan per Musim")
        fig_s, ax_s = plt.subplots(figsize=(8, 5))
        sns.barplot(x='season', y='cnt', data=df_filtered, estimator=np.mean, palette='Set2', ax=ax_s)
        ax_s.set_xticklabels(['Spring', 'Summer', 'Fall', 'Winter'])
        ax_s.set_xlabel("Musim")
        ax_s.set_ylabel("Rata-rata Jumlah Sewa")
        st.pyplot(fig_s)

        # Logika Peningkatan yang kamu buat
        season_avg = df_filtered.groupby('season')['cnt'].mean()
        if 1 in season_avg and 3 in season_avg:
            peningkatan = ((season_avg[3] - season_avg[1]) / season_avg[1]) * 100
            st.write(f"**Info:** Peningkatan Musim Panas (Fall) vs Musim Dingin (Spring) adalah **{peningkatan:.1f}%**")

    with col2:
        st.subheader("Pengaruh Kondisi Cuaca")
        fig_w, ax_w = plt.subplots(figsize=(8, 5))
        sns.boxplot(x='weathersit', y='cnt', data=df_filtered, palette='Set3', ax=ax_w)
        ax_w.set_xticklabels(['Clear', 'Mist', 'Light Rain/Snow'], errors='ignore')
        ax_w.set_xlabel("Kondisi Cuaca")
        st.pyplot(fig_w)

# --- TAB 2: REGISTERED USERS & KORELASI ---
with tab2:
    st.header("Analisis Registered Users & Jam Puncak")

    # Visualisasi Registered Users per Jam
    st.subheader("Rata-rata Registered Users per Jam (Working Day vs Libur)")
    fig_reg, ax_reg = plt.subplots(figsize=(12, 6))
    sns.lineplot(x='hr', y='registered', hue='workingday', data=df_filtered, marker='o', ax=ax_reg)
    ax_reg.set_xticks(range(0, 24))
    ax_reg.set_xlabel("Jam (hr)")
    ax_reg.set_ylabel("Rata-rata Registered Users")
    ax_reg.grid(True)
    st.pyplot(fig_reg)

    # Metrik Bisnis Berdasarkan Analisis Kamu
    st.divider()
    col_a, col_b = st.columns(2)

    with col_a:
        st.subheader("Korelasi Antar Variabel")
        fig_heat, ax_heat = plt.subplots(figsize=(10, 8))
        corr_matrix = df_filtered[['cnt','temp','hum','windspeed','casual','registered']].corr()
        sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', ax=ax_heat)
        st.pyplot(fig_heat)

    with col_b:
        st.subheader("Estimasi Potensi Bisnis")
        # Menghitung jam puncak working day
        hourly_reg = df_filtered[df_filtered['workingday'] == 1].groupby('hr')['registered'].mean()
        peak_hr = hourly_reg.idxmax()
        avg_peak = hourly_reg.max()

        additional_bikes = avg_peak * 0.20
        revenue_est = additional_bikes * 365 * 3 # asumsi $3/jam

        st.metric("Jam Puncak (Working Day)", f"Jam {peak_hr:02d}")
        st.metric("Saran Tambahan Armada (20%)", f"{additional_bikes:.0f} Sepeda")
        st.metric("Estimasi Potensi Revenue (1 Thn)", f"${revenue_est:,.0f}")
        st.info("💡 Insight: Fokuskan distribusi sepeda pada jam masuk/pulang kantor.")

# --- TAB 3: ANALISIS RFM ---
with tab3:
    st.header("Segmentasi Hari Berdasarkan RFM")

    # Menyiapkan data RFM (Daily)
    # Kita groupby dteday untuk mendapatkan performa harian
    rfm_df = df_all.groupby('dteday').agg({
        'registered': 'sum'
    }).reset_index()

    # Recency
    max_date = rfm_df['dteday'].max()
    rfm_df['Recency'] = (max_date - rfm_df['dteday']).dt.days

    # Frequency & Monetary
    rfm_df['Frequency'] = rfm_df['registered']
    rfm_df['Monetary'] = rfm_df['registered'] * 3 # Asumsi $3 per sewa

    # Scoring
    rfm_df['R_Score'] = pd.qcut(rfm_df['Recency'], 5, labels=[5,4,3,2,1])
    rfm_df['F_Score'] = pd.qcut(rfm_df['Frequency'], 5, labels=[1,2,3,4,5])
    rfm_df['M_Score'] = pd.qcut(rfm_df['Monetary'], 5, labels=[1,2,3,4,5])

    rfm_df['RFM_Score'] = rfm_df['R_Score'].astype(int) + rfm_df['F_Score'].astype(int) + rfm_df['M_Score'].astype(int)

    def rfm_segment(score):
        if score >= 12: return 'Champion'
        elif score >= 9: return 'Loyal Customer'
        elif score >= 6: return 'Potential'
        else: return 'At Risk'

    rfm_df['Segment'] = rfm_df['RFM_Score'].apply(rfm_segment)

    # Visualisasi RFM
    col_c, col_d = st.columns([1, 1])

    with col_c:
        st.subheader("Data Table RFM")
        st.dataframe(rfm_df[['dteday', 'Recency', 'Frequency', 'Monetary', 'Segment']].head(10))

    with col_d:
        st.subheader("Distribusi Segmen Hari")
        segment_df = rfm_df['Segment'].value_counts().reset_index()
        segment_df.columns = ['Segment', 'Count']
        fig_pie = px.pie(segment_df, values='Count', names='Segment', hole=0.4,
                         color_discrete_sequence=px.colors.qualitative.Pastel)
        st.plotly_chart(fig_pie, use_container_width=True)

    st.table(rfm_df.groupby('Segment').agg({
        'Recency': 'mean',
        'Frequency': 'mean',
        'Monetary': 'mean'
    }).round(1))
```
